```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.tsv', sep='\t')
test = pd.read_csv('data/valid.tsv', sep='\t')
```

```python
text = list(train.Post.values)
text += list(test.Post.values)
```

```python
len(text)
```

```python
with open('texts.txt', 'w+') as f:
    f.write('\n'.join(text))
```
